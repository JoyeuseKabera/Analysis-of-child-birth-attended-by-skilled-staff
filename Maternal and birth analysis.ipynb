###  Data processing

After loading the dataset(the birth attended by skilled stuff), I filtered to remove the years which doesn't have information for many countries 
because there is no other better way to handle this kind of nan in a dataset

```python

#Drop column to start from 1990
skill_Birth = skill_Birth.drop(['1960', '1961','1962','1963','1964','1965','1966','1967','1968','1969',
                  '1970','1971','1972','1973','1974','1975','1976','1977','1978',
                  '1979','1980','1981','1982','1983','1984','1985','1986','1987','1988','1989'], axis=1)
skill_Birth
```

###### Choosing the latest year which contains much info compared to other years

```python
columns = ['Country Name','Country Code','2017']
df1 = pd.DataFrame(skill_Birth, columns=columns)
df1
```

```python
## droping the Nan values
df1 = df1.dropna()
df1.index = np.arange(0,83,1)
df1
```

```python
df1[df1['Country Code']=='HTI']
```

```python
print("The minimum is", df1['2017'].min())
print("The maximum is", df1['2017'].max())
```

```python
df1[df1['Country Code']=='FIN']
```

###### The plot demonstrate the number of skilled staff in different countries around the world

```python

fig = plt.figure(figsize = (15, 7))
y = df1['2017']
x = df1.index
n = df1['Country Code']
#plt.scatter(x, y, color ='green')

plt.scatter(x, y)

for i, txt in enumerate(n):
    plt.annotate(txt, (x[i], y[i]))
    
plt.xlabel("Index of the country")
plt.ylabel("No. of birth attended by skilled stuff")
plt.title("birth attended by skilled stuff in 2017")
plt.show()
```

##### The graph show number of skilled staff ranging from 41.6 of Haiti to around 100 of Finland, 
##### Developping countries tend to have a low number of skilled staff compared to developped countries.

##### Droping the country which do not conntain the data for those selected years

```python
df2 = df2.dropna()
df2
```

##### bar plot comparing the number attended by skilled stuff in different countries

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (15, 8))
X_axis = np.arange(len(df2['Country Name']))
  
plt.bar(X_axis - 0.2, df2['1990'], 0.4, label = '1990')
plt.bar(X_axis + 0.2, df2['2017'], 0.4, label = '2017')
  
plt.xticks(X_axis, df2['Country Name'])

plt.legend()
plt.xlabel("Country name")
plt.xticks(rotation = 45)
plt.ylabel("No. of birth attended by skilled stuff")
plt.title("birth attended by skilled stuff in 2017")
plt.show()
```

##### The above figure shows that there is an improvement in number of birth attended by skilled staffs

## Analysing maternal mortality versus GDP per capita

```python
maternalMortality_vs_gdp = pd.read_csv("MaternalMortalityRatio-vs-gdp-per-capita.csv") 
maternalMortality_vs_gdp
```

```python
#Drop unwanted columns
maternalMortality_vs_gdp.drop(maternalMortality_vs_gdp.columns[[1, 6]], axis = 1, inplace = True)
maternalMortality_vs_gdp
```

```python
maternalMortality_vs_gdp = maternalMortality_vs_gdp.dropna()
maternalMortality_vs_gdp
```

#### Analyzing data in Rwanda

```python
df9_Rwanda = maternalMortality_vs_gdp.loc[maternalMortality_vs_gdp['Entity'] == 'Rwanda']
df9_Rwanda
```

```python
df9_Rwanda.plot(x="Year", y=["Maternal Mortality Ratio (Gapminder (2010) and World Bank (2015))", 
                     "GDP per capita, PPP (constant 2017 international $)"], kind="bar")
plt.title("Maternal Mortality Ratio vs GDP per capita")
plt.show()
```

#### The plot demonstrate that the increase in GDP from 1990 to 2015 with exception of 1994 where the GDP decreased due to Genocide.
#### contrally, the maternal mortality decreases from 1990 to 2015

#### From different analysed datasets , the number of mortality rates are reduced with the increase in skilled staff in health care; therefore, increasing GDP per capita and investing in education especially in health care field will be fruitfull to both government and citizens.
